使用运算函数，可以预处理Nan值

```python
# 14天的四种蔬菜的价格
datas = np.random.randint(1, 10, size=(14, 4))
columns = ('白萝卜', '西红柿', '黄瓜', '白菜')
days = np.arange(1, 15, step=1)

df = DataFrame(datas,days,columns)
df
```

```python
# 将所有蔬菜的价格上调1.5倍
df *= 2.5
df
```

```python
# 将前5天的所有蔬菜价格打8.8折处理
df[:5] *= 0.88
df
```

```python
# 将第7,9,11三天的白萝卜价格降半价，而且减1元
df.loc[7:12:2, '白萝卜'] = df.loc[7:12:2, '白萝卜']/2 -1
df
```

```python
df.iloc[[1, 5, 8], [0, 3]] = np.nan
df
```

```python
df.add(1, fill_value=0)
```

```python
df2 = DataFrame({
    'A': [1, 2, 3],
    'B': [5, 6, 7],
    'C': [8, 9, 10]
})
df2
```

```python
# axis = 'index'  ==> axis = 1
df2.add([1, 2, 3], axis='columns')  # 按列相加，每一行的数值按照列的顺序相加
```

```python
# axis = 'index'  ==> axis = 0
df2.add(Series([9, 10, 12]), axis='index')   # 按行相加，每列按行顺序相加
```

```python
# 增加第四列，折扣列（8.8 - 5.5）
rate = np.round(Series(np.random.uniform(0.55, 0.89, size=3)), 2)
rate
```

```python
# 新增一列
df2['rate'] = rate
df2
```

```python
# 将B列的数值和rate列的数值相乘，并将结果生成result列
df2['result'] = df2['B'].mul(df2['rate'])  # 两个Series按行索引进行乘法运算
df2
```

假设张三期中考试数学被发现作弊，要记为0分，如何实现？
李四因为举报张三作弊立功，期中考试所有科目加100分，如何实现？
后来老师发现有一道题出错了，为了安抚学生情绪，给每位学生每个科目都加10分，如何实现？

```python
names = ['张三', '李四', '王五', '赵六']
course = ['语文', '数学', '英语']
# 其中考试
ddd = DataFrame(np.random.randint(100, size=(4, 3)), names, course)
ddd
```

```python
# 期末考试
ddd2 = DataFrame(np.random.randint(100, size=(4, 3)), names, course)
ddd2
```

```python
(ddd + ddd2)/2
```

```python
ddd.add(ddd2).div(2)
```

```python
# 假设张三其中考试数学发现作弊， 要记为0分，如何实现？
ddd.loc['张三', '数学'] = 0
ddd
```

```python
# 李四因为举报张三作弊立功，其中考试所有科目加100分，如何实现/
ddd.loc['李四'] += 100
ddd
```

```python
# 后来老师发现有一道题出错了，为了安抚学生情绪，给每位学生每个科目都加10分，如何实现？
ddd.add(10)
```
