### 什么是透视表
- 透视表针对特定的列进行分组同时聚合计算，相当于groupby()与聚合函数的组合使用
- 透视表注重的是数据的排列与聚合计算，默认使用平均值做聚合计算

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

### 行透视表
- df.pivot_table(index=)

```python
df = DataFrame({
    'sex': np.random.choice(['男', '女'], size=10),
    'smoke': np.random.choice(['Yes', 'No'], size=10),
    'height': np.random.randint(160, 200, size=10),
    'weight': np.random.randint(55, 120, size=10),
    'size': np.random.randint(30, 40, size=10),
})
df
```

```python
# 按是否抽烟，计算平均身高、体重、三围
df.groupby('smoke').mean()
```

```python
df.pivot_table(index='smoke', aggfunc='mean')
```

```python
# 查看是否抽烟和不同的性别的平均的身高、体重、三围的数据
df.groupby(['smoke', 'sex']).mean()
```

```python
df.pivot_table(index=['smoke', 'sex'], aggfunc='max')
```

### 列透视表
- df.pivot_table(columns=)

```python
df.pivot_table(columns='smoke').transpose()
```

```python
df.groupby(['smoke']).mean().transpose()
```

```python
df.pivot_table(columns=['smoke', 'sex'])
```

### 行列同时透视
- df.pivot_table(index, columns)

```python
df.pivot_table(index='sex', columns='smoke', aggfunc='min')
```

### 交叉表
- pd.crosstab(index, columns) 通过不同表的特征数据进行交叉显示
- 是一种特殊的透视表,聚合函数只是统计数量

```python
# margins=True 小计
pd.crosstab(df['sex'], df['smoke'], margins=True, margins_name='小计')
```
