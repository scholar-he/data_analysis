```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
moon= plt.imread('moonlanding.png')  # 返回ndarray
plt.imshow(moon)
```

```python
moon.shape
```

```python
# 灰度显示
plt.imshow(moon, cmap='gray')
```

```python
moon[:2, :2]  # png格式的图片的内容，数值是小数位
```

```python
# 图片消噪，傅里叶变换可以实现
from scipy.fftpack import fft2, ifft2
import numpy as np
```

```python
# 将时域的数据转成频域的数据
fft_moon = fft2(moon)

# 找出高频次的数据，并置位0 -> 去除高频词数据
fft_moon = np.where(np.abs(fft_moon) > 8e2, 0, fft_moon)

# 将频域的数据转成时域的数据,并提取复数中的实数部分
ifft_moon = np.real(ifft2(fft_moon))

plt.imshow(ifft_moon, cmap='gray')
```

```python
from scipy.misc import face
```

```python
face_img = face()  # gray=True 表示返回的图片已被灰度处理
```

```python
plt.imshow(face_img)
```

```python
face_img.shape
```

```python
face_img[:5, :5]  # 三个元素表示RGB的0-255的颜色值，图片的格式是jpg
```

```python
# 最大化或最小化或平均值等方式灰度处理
plt.imshow(face_img.max(axis=-1), cmap='gray')
```

```python
plt.imshow(face_img.min(axis=-1), cmap='gray')
```

```python
face_img.mean(axis=-1).shape
```

```python
plt.imshow(face_img.mean(axis=-1), cmap='gray')
```
