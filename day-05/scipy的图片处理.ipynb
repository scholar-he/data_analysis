```python
# 
from scipy.fftpack import fft2, ifft2
import numpy as np
```

```python
# 
fft_moon = fft2(moon)

# 0 -> 
fft_moon = np.where(np.abs(fft_moon) > 8e2, 0, fft_moon)

# ,
ifft_moon = np.real(ifft2(fft_moon))

plt.imshow(ifft_moon, cmap='gray')
```
