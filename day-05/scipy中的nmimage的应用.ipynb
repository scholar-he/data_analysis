### 图片的移动缩放和旋转
- ndimage.shift()  移动
- ndimage.zoom()   缩放
- ndimage.rotate()  旋转

```python
import numpy as np
from scipy.misc import face
from scipy.ndimage import shift, zoom, rotate
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
face_img = face(gray=True)
plt.imshow(face_img, cmap='gray')
```

```python
# mode = {'reflect', 'constant', 'nearest', 'mirror', 'wrap'}
plt.imshow(shift(face_img, shift=(150, 100), mode='mirror'), cmap='gray')
```

```python
plt.imshow(rotate(face_img, angle=45, mode='nearest'), cmap='gray')
```

```python
face_r = rotate(face_img, angle=45, mode='nearest')
plt.imsave('face_rotate.png', face_r)
```

```python
face_2 = zoom(face_img, zoom=0.5)
plt.imshow(face_2, cmap='gray')
```

```python
plt.figure(figsize = (12,12))  # 设置画布大小

faces = (
    face_img,
    shift(face_img, shift=(150, 150)),
    rotate(face_img, angle=90),
    zoom(face_img, zoom=0.5)
)

for i,face in enumerate(faces):
    plt.subplot(1, 5,  i+1)  # 创建子画布
    plt.imshow(face,cmap = plt.cm.gray )  # 在子画布中显示图片
    plt.axis('off')  # 将坐标轴关闭
plt.show()
```
